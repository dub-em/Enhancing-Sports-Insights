```python
options = Options()
#options.headless = True
ua = UserAgent()
userAgent = ua.random
options.add_argument(f'user-agent={userAgent}')


driver = webdriver.Chrome(r'C:\Users\HP\Downloads\Premier_League_Stats\Loading_with_selenium\chromedriver.exe',options=options)


#sportsloci
premier_league = 'https://sportsloci.com/league/Premier-League/1204'
la_liga = 'https://sportsloci.com/league/La-Liga/1399'
seria_a = 'https://sportsloci.com/league/Serie-A/1269'
primeria = 'https://sportsloci.com/league/Primeira-Liga/1352'

def get_league_matches(league_url):
    driver.get(league_url)
    contents =driver.find_elements_by_class_name("upcoming_games_list")
    content = [con.get_attribute('innerText') for con in contents]

    match_links = driver.find_elements_by_css_selector("div ul li a")
    match_link = [match.get_attribute('href') for match in match_links]
    fixtures = [url for url in match_link if 'fixture' in url]
    clean = [item.replace('\n', ',').split(',') for item in content]
    clean = list(chain(*clean))
    data = [x for x in clean if not x.startswith('Week')]
    data_chunks = [data[i:i+5] for i in range(0, len(data), 5)]
    columns = ['Date', 'Time', 'Home Team', 'vs', 'Away Team']
    df = pd.DataFrame(data_chunks, columns=columns)
    df['links'] = fixtures
    driver.quit()
    
    def get_match_stats(match_url):
        options = Options()
        #options.headless = True
        ua = UserAgent()
        userAgent = ua.random
        options.add_argument(f'user-agent={userAgent}')


        driver = webdriver.Chrome(r'C:\Users\HP\Downloads\Premier_League_Stats\Loading_with_selenium\chromedriver.exe',options=options)

        driver.get(match_url)
        contents =driver.find_elements_by_class_name("head_to_head_list")
        content = [con.get_attribute('innerText') for con in contents]

        clean = [item.replace('\n', ',') for item in content]
        data_list = [item.split(',') for item in clean]
        flat = [element for innerList in data_list for element in innerList]

        # Find the index of 'Games'
        data = flat[3:]
        teams = flat[:2]

        num_metrics = len(data) // 3
        columns = ['Metric', teams[0], teams[1]]
        rows = [data[i:i+3] for i in range(0, len(data), 3)]

        # Create a Pandas DataFrame
        df = pd.DataFrame(rows, columns=columns)
        df.loc[len(df.index)] = ['based', 'home', 'away'] 

        driver.quit()
        print(df)
    
    for i in df.links:
        get_match_stats(i)
    return df

get_league_matches(premier_league)
```

```python
options = Options()
#options.headless = True
ua = UserAgent()
userAgent = ua.random
options.add_argument(f'user-agent={userAgent}')


driver = webdriver.Chrome(r'C:\Users\HP\Downloads\Premier_League_Stats\Loading_with_selenium\chromedriver.exe',options=options)

driver.get('https://sportsloci.com/fixture/Brentford-Crystal_Palace/3330477')
contents =driver.find_elements_by_css_selector("div div div table")
content = [con.get_attribute('innerText') for con in contents]
driver.quit()
clean = [item.replace('\t', ' ') for item in content]
clean = [item.replace('\n', ',').split(',') for item in clean]

clean = list(chain(*clean))
clean = clean[1:]
cleaned = [' '.join(line.split(' Premier League')[0:-1]) for line in clean]

positions = []
teams = []
played = []
won = []
drawn = []
lost = []
goals_for = []
goals_against = []
goal_difference = []
points = []

for entry in cleaned:
    parts = entry.split()
    positions.append(parts[0])
    teams.append(' '.join(parts[1:-8]))
    played.append(parts[-8])
    won.append(parts[-7])
    drawn.append(parts[-6])
    lost.append(parts[-5])
    goals_for.append(parts[-4])
    goals_against.append(parts[-3])
    goal_difference.append(parts[-2])
    points.append(parts[-1])

# Creating the DataFrame
df = pd.DataFrame({
    'Position': positions,
    'Team': teams,
    'Played': played,
    'Won': won,
    'Drawn': drawn,
    'Lost': lost,
    'Goals For': goals_for,
    'Goals Against': goals_against,
    'Goal Difference': goal_difference,
    'Points': points
})

df
```
